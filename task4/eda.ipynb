```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data

df = pd.read_csv('drive/MyDrive/globalterrorismdb.csv', encoding='ISO-8859-1')

df.head()
```

```python
df.describe()
```

```python
# Drop all rows with NaN values in any column
df = df.dropna(axis = 1)

# Display the cleaned DataFrame
df.head()
```

```python
# Incidents by country
country_counts = df['country_txt'].value_counts().head(20)  # Top 20 countries
plt.figure(figsize=(15, 8))
country_counts.plot(kind='bar')
plt.title('Top 20 Countries with Most Terrorism Incidents')
plt.xlabel('Country')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()

# Incidents by region
region_counts = df['region_txt'].value_counts()
plt.figure(figsize=(12, 6))
region_counts.plot(kind='bar')
plt.title('Number of Terrorism Incidents by Region')
plt.xlabel('Region')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()
```

```python
attack_counts = df['attacktype1_txt'].value_counts()
plt.figure(figsize=(12, 6))
attack_counts.plot(kind='bar')
plt.title('Types of Terrorism Attacks')
plt.xlabel('Attack Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby('iyear').size().plot()
plt.title('Number of Terrorism Incidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()
```

```python
success_counts = df['success'].value_counts()
plt.figure(figsize=(6, 6))
success_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Success of Terrorism Incidents')
plt.show()

suicide_counts = df['suicide'].value_counts()
plt.figure(figsize=(6, 6))
suicide_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Suicide Attacks')
plt.show()
```

```python
# Example: Correlation matrix
correlation_matrix = df[['success', 'suicide', 'property']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
pivot_table = df.pivot_table(index='iyear', columns='region_txt', values='eventid', aggfunc='count', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True)
plt.title('Heatmap of Terrorism Incidents by Region and Year')
plt.show()
```
